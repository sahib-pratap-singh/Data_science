### Curious case of 'for' loops 

##### It is very important to understand for loops. 

```python
# Defining a new list:
my_list=['I','Love','Data','Science']
```

##### We want to use for loop and print each element of out list 'my_list'

```python
for i in my_list:
    print(i)
```

##### Now, We want elements at even index from out list 'my_list'

i.e. I and data

We have to use index in this case.

To get total length we know, len() can be used.

```python
for i in len(my_list):
    print(i)
```

#### Understand the error, for loop needs a starting point, end point and jumping interval. 
#### What we had done above is defining only the end point.
#### We must define that it should start from 0 to len(my_list) and jump 1 unit each time.

##### This is where we must understand range function

```python
# As the name suggests range fucntion gives us the range of numbers
for i in range(0,10):   # this will give 0 to 10 with 1 unit interval
    print(i,end='')
    
# Note last number is always excluded.
```

```python
for i in range(0,10,2):   # this will give 0 to 10 with 2 unit interval
    print(i)
```

##### Now, We want elements at even index from out list 'my_list'

```python
# complete the code and run it:
for i in range(0,len(my_list),2): # this will give 0 to length of my_list with 2 unit interval
    print(    )
    
```

#### If we want each element

```python
for i in my_list:
    print(i)
```

#### If we want index of each element

```python
for i in range(0,len(my_list),1):
    print(i)
```
