### What is dataframe?

#### It is a tabular representation of data

```python
# Let's take 'df_marks' as the name of our dataframe (choosing names is our choice, it could be ANYTHING)

df_marks=pd.DataFrame(list_marks)

# pd.DataFrame() is the function which creates dataframes (as the name also suggests)
# pandas.DataFrame() is the same thing.
# if you have done something like: 
# import pandas as i_love_data_science
# i_love_data_science.DataFrame() can also be used


# We passed our list which will be converted into tabular form


# Let's see our dataframe
df_marks
```

##### Looks pretty than list_marks, doesn't it.

#### We want to add a column of percentage (as we did in our list_marks)

### Always consider dataframe as mutliple series (quite same as list, but different!)

 Therefore, functions you will perform let's say sum() function, 

it will be implemented column-wise (not row-wise, you have to specify)

```python
# Finding mean
df_marks.mean()
```

##### We can see that we got mean column-wise.
##### But to find percentage of each student we want mean row-wise.
##### It is quite simple,
##### while calling function if we say 
##### axis=1 (it means row-wise)
##### axis=0 (it means column-wise)

### Remember using 'Shift + Tab'

##### If we use 'shift+Tab' on our mean function above this is what we will get:

#### From now on we won't call it row-wise but index-wise (as said by pandas itself above)

```python
# Finding mean index-wise
df_marks.mean(axis=1)
# This is the percentage of each student.
```

#### We want to make a fourth column for percentage.

#### While making a new column in dataframes just imagine that it already exists

```python
# We want to call our column as 'Percentage'
df_marks['Percentage']=df_marks.mean(axis=1)
```

```python
df_marks
```
