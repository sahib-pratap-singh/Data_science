```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
unemp=pd.read_csv('C:\\Users\\sahib pratap\\Downloads\\quarantine time\\unemp.csv')
```

```python
unemp['TIME']=pd.to_datetime(unemp['TIME'])
```

### We are using pivot tables:

```python
df1=unemp.pivot_table(
    values='Value',
    index='TIME',
    columns='LOCATION'
)


# Plot:
df1.plot(legend=False)
plt.xlim(pd.Timestamp('2000-01-1'), pd.Timestamp('2020-03-01'))
#plt.ylim(3,10)    # You can change y limit as well
```

#### We want to plot values for Australia and USA only

```python
df1.plot(y=['AUS','USA'])
```
