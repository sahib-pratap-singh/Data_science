#### This is a better method and would be much useful in future.

#### This is the output we want:

```python
c=pd.DataFrame({'LOCATION':[],"TIME":[],"MEN":[],"WOMEN":[]})
```

```python
country=list(w_unemp[w_unemp['TIME']>='2020']['LOCATION'].unique())
country
```

```python
for place in country:
    value_men=w_unemp[ (w_unemp['SUBJECT']=='MEN') & (w_unemp['TIME']>'2020') & (w_unemp['LOCATION']==place)][['TIME','Value']]
    value_women=list(w_unemp[ (w_unemp['SUBJECT']=='WOMEN') & (w_unemp['TIME']>'2020')&(w_unemp['LOCATION']==place)]['Value']) 
    for i in range(len(value_women)):
        c=c.append({'LOCATION':place,'TIME':value_men.iloc[i,0],'MEN':value_men.iloc[i,1],'WOMEN':value_women[i]},ignore_index=True)        
        
c.rename(columns={'LOCATION':'Location','TIME':'Time'})
```

```python
gdf = w_unemp.loc[(w_unemp['TIME'].str.contains('2020'))].groupby('SUBJECT')
```

 #### Logic:

#### To see what we have in gdf: 

```python
for i,g in gdf: 
    print(i)
    print('\n')  # To have gap
    print(g)
    print('\n')   # To have gap
```

```python
sub = [] # List in which we will save subject
val=[]  # List in which we will save values
locn=[]  # List in which we will save location
t=[] # List in which we will save time

for i,g in gdf:                       # as we did earlier
    sub.append(i)  
    val.append(list(g['Value']))
    locn.append(list(g['LOCATION'])) 
    t.append(list(g['TIME'])) 
               
```

#### Note: It is very important that we use append function. Otherwise it will give completely absurd output

```python
sub
```

```python
val
```

```python
t 
```

```python
new_df=pd.DataFrame() # Creating new dataframe

for i in range(len(sub)):         # we are saying go from 0 to length of sub list i.e. 3
    new_df[sub[i]] = val[i]       # We save column with column-name = subject (men, women and total) and values in it as 'values' 
    new_df['Location'] = locn[i]  # We save location in column named location
    new_df['Time']=t[i]           # We save time in column named 'Time'
```

```python
new_df
```

```python
# We don't want to see total and we want to change the sequence of columns

new_df=new_df[['Location','Time','MEN','WOMEN']]
new_df
```
